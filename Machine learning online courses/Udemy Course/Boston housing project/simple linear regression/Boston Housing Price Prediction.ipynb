# Linear Regression with Scikit-Learn

```python
X2=df.loc[:,df.columns=='RM']
X2=X2.values.reshape(-1,1)
X2
```

```python
X1 = df['RM'].values.reshape(-1,1)
X1
```

```python
y = df['MEDV'].values
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X1, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,10))
ax = sns.regplot(x='RM',y='MEDV',data=df)
ax.set(xlabel='common xlabel', ylabel='common ylabel')
plt.show();
```

```python
plt.figure(figsize=(8,10))
ax = sns.jointplot(x='RM', y='MEDV',kind = 'reg', data=df,size=10);
ax.set_axis_labels('x', 'y', fontsize=16)
plt.show();
y1=model.predict(X1)
```

# MODEL EVALUATION

# mean_squared error

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(X1,y,train_size = 0.8)
model.fit(X_train,y_train)
pred = model.predict(X_test)

mean_squared_error(y_test,pred)
```

# COEFFICIENT OF DETERMINATION, R^2 = 1- SSE(sum of squared errors)/SST (total sum of squares)

```python
from sklearn.metrics import r2_score
#the higher this score the best
r2_score(y_test,pred)
```

**ROBUST REGRESSION**


```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X1,y)
```

```python
inliers = ransac.inlier_mask_
outliers = np.logical_not(inliers)
```

```python
pred=ransac.predict(X1)
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(20,10))
plt.scatter(X1[inliers],y[inliers],color='blue', marker ='s', label='inliers')
plt.scatter(X1[outliers],y[outliers],color='red', marker ='s', label='outliers')
plt.plot(X1,pred,linestyle='-',color='pink')
plt.plot(X1,y1,linestyle='-',color='blue')
plt.legend(loc='upper left')
plt.show
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

# Performance evaluation of linear regression model 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1,y, train_size=0.8)
from sklearn.metrics import mean_squared_error
ransac.fit(X_train,y_train)
pred=ransac.predict(X_test)
mean_squared_error(y_test,pred)
```
