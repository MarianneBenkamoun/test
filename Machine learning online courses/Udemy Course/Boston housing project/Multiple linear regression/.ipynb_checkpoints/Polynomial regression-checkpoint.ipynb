```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
```

```python
df =pd.read_csv('../housing.data',delim_whitespace=True,header=None)
```

```python
col_names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
df.columns = col_names
X = df.loc[:,df.columns.isin(['CRIM','INDUS','NOX','RM','AGE','DIS','PTRATIO','LSTAT'])]
y = df['MEDV'].values
y
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
col_X = X.columns
X_std = pd.DataFrame(X_std)
X_std.columns = col_X
```

```python
import seaborn as sns
for col in X_std.columns: 
    plt.figure(figsize=(10,10))
    sns.boxplot(X_std[col])
```

```python
import seaborn as sns
for col in X.columns: 
    plt.figure(figsize=(10,10))
    plt.scatter(X[col],y,label=col) 
    plt.xlabel(col)
    plt.ylabel('price');
```

```python
import seaborn as sns
for col in X_std.columns: 
    plt.figure(figsize=(10,10))
    plt.scatter(X_std[col],y,label=col) 
    plt.xlabel(col)
    plt.ylabel('price');
```

```python
X_train_std = scaler.fit_transform(X_train)
X_train_std = pd.DataFrame(X_train_std)
X_train_std.columns = col_X
X_test_std = scaler.transform(X_test)
X_test_std = pd.DataFrame(X_test_std)
X_test_std.columns = col_X
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.metrics import r2_score
lr.fit(X_train_std,y_train)
pred = lr.predict(X_test_std)

r2_score(y_test, pred)  
```

```python
from sklearn.preprocessing import PolynomialFeatures

def create_polynomial_regression_model(degree):
  "Creates a polynomial regression model for the given degree"
  
  poly_features = PolynomialFeatures(degree=degree)
  
  # transforms the existing features to higher degree features.
  X_train_poly = poly_features.fit_transform(X_train_std)
  
  # fit the transformed features to Linear Regression
  poly_model = LinearRegression()
  poly_model.fit(X_train_poly, y_train)
  
  # predicting on training data-set
  y_train_predicted = poly_model.predict(X_train_poly)
  
  # predicting on test data-set
  y_test_predict = poly_model.predict(poly_features.fit_transform(X_test_std))
  
  # evaluating the model on training dataset
  rmse_train = np.sqrt(mean_squared_error(y_train, y_train_predicted))
  r2_train = r2_score(y_train, y_train_predicted)
  
  # evaluating the model on test dataset
  rmse_test = np.sqrt(mean_squared_error(y_test, y_test_predict))
  r2_test = r2_score(y_test, y_test_predict)
  
  print("The model performance for the training set")
  print("-------------------------------------------")
  print("RMSE of training set is {}".format(rmse_train))
  print("R2 score of training set is {}".format(r2_train))
  
  print("\n")
  
  print("The model performance for the test set")
  print("-------------------------------------------")
  print("RMSE of test set is {}".format(rmse_test))
  print("R2 score of test set is {}".format(r2_test))
```

```python
create_polynomial_regression_model(2)
```
