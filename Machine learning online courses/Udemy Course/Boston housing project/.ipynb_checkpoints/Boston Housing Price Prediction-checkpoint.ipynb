**ROBUST REGRESSION**


```python
X=df['RM'].values
X = X.reshape(-1,1)
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X,y)
```

```python
inliers = ransac.inlier_mask_
outliers = np.logical_not(inliers)
```

```python
pred=ransac.predict(X)
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(20,10))
plt.scatter(X[inliers],y[inliers],color='blue', marker ='s', label='inliers')
plt.scatter(X[outliers],y[outliers],color='red', marker ='s', label='outliers')
plt.plot(X,pred,linestyle='-',color='pink')
plt.plot(X,y1,linestyle='-',color='blue')
plt.legend(loc='upper left')
plt.show
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```
