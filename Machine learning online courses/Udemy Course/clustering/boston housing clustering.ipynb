```python
import pandas as pd
df = pd.read_csv('../Boston housing project/housing.data',delim_whitespace=True,header=None)
col_names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
df.columns = col_names
X = df.loc[:,df.columns!='MEDV']
col_X = X.columns 
y = df['MEDV'].values
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import itertools
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestRegressor
from mlxtend.classifier import EnsembleVoteClassifier
from mlxtend.plotting import plot_decision_regions

# Initializing Classifiers
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=9)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
for i in y_kmeans:
    print(i)
    
```

```python
# in row_dict we store actual meanings of rows, in my case it's russian words
clusters = {}
n = 0
for item in y_kmeans:
    if item in clusters:
        clusters[item].append(X.iloc[n,:])
    else:
        clusters[item] = [X.iloc[n,:]]
    n +=1
for item in clusters:
    print ("Cluster ", item)
    for i in clusters[item]:
        print(i)
```
